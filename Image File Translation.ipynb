# 이미지에서 텍스트 출력

```python
from PIL import Image
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
text = pytesseract.image_to_string(Image.open('eng.jpeg'), lang = "eng")
text1 = text.replace("\n", " ")

print(text)
print(text.replace("\n", " ")) # 문장 이어붙이기
```

# 출력된 텍스트를 번역

```python
import googletrans
translator = googletrans.Translator()

result = translator.translate(text1, dest='ko')

result.text
```
